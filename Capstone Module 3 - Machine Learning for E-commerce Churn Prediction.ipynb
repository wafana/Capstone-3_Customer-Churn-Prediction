### **1.4.3. Evaluasi Model**
Evaluation metric dalam machine learning digunakan untuk mengukur kinerja model dalam melakukan prediksi. Pemilihan metric bergantung pada jenis masalah yang dihadapi.



Pada kasus ini, berikut adalah definisi 0 dan 1 pada label yang akan diprediksi:
- Tidak Churn : 0
- Churn : 1

Maka interpretasinya hasil aktual dan prediksi adalah:
- True Positive : Aktual Churn, Prediksi Churn
- True Negative : Aktual Tidak Churn, Prediksi Tidak Churn
- False Positive : Aktual Tidak Churn, Prediksi Churn
- False Negative : Aktual Churn, Prediksi Tidak Churn

Presisi dan recall:
- **Presisi:** Proporsi prediksi positif yang benar. Formula: TP / (TP + FP).
- **Recall:** Proporsi kasus positif yang berhasil terdeteksi. Formula: TP / (TP + FN).
- **F2-score:** Memberi bobot lebih besar pada recall. Ini berguna ketika false negative (FN) lebih berbahaya daripada false positive (FP). Skor F2 membantu menemukan keseimbangan antara precision dan recall, dengan fokus utama pada mengurangi FN.
- 
Dalam prediksi churn, **F2 score** membantu model lebih fokus pada recall, yaitu menangkap sebanyak mungkin pelanggan yang benar-benar akan berhenti (**false negative**). Jika terlalu banyak pelanggan churn yang terlewat, bisnis bisa kehilangan peluang untuk mencegah mereka pergi, misalnya dengan penawaran khusus atau peningkatan layanan.  

Di sisi lain, jika model salah menandai pelanggan yang sebenarnya tetap bertahan (**false positive**), dampaknya lebih kecil—mungkin hanya biaya insentif yang tidak perlu. Karena itu, **F2 score** lebih cocok dibandingkan metrik seperti F1 score, karena memberi bobot lebih besar pada recall daripada precision.

Sumber: [Evaluation Metric](https://telnyx.com/learn-ai/f2-score)

> - `WarehouseToHome`
>   - Missing Values : Diisi menggunakkan Simple Imputer dengan metode median, karena distribusi data pada kolom ini tidak normal.
>   - Scaling : Menggunakan Robust Scaler karena distribusi memiliki extreme outliers.
> - `NumberOfDeviceRegistered`
>   - Scaling : Menggunakan Min Max Scaler karena distribusi data tidak normal dan tidak terdapat extreme outliers.
> - `PreferedOrderCat`
>   - Encoding : Menggunakan One Hot Encoding karena nilai unik pada kolom masih cenderung sedikit (>10).
> - `SatisfactionScore`
>   - Scaling : Menggunakan Min Max Scaler karena distribusi data tidak normal dan tidak terdapat extreme outliers.
> - `MaritalStatus`
>   - Encoding : Menggunakan One Hot Encoding karena nilai unik pada kolom masih cenderung sedikit (>10).
> - `NumberOfAddress`
>   - Scaling : Menggunakkan Robust Scaler karena distribusi memiliki extreme outliers.
> - `Complain`
>    - Scaling : Menggunakan Min Max Scaler karena distribusi data tidak normal dan tidak terdapat extreme outliers.

>   - Missing Values : Diisi menggunakkan Simple Imputer dengan metode median, karena distribusi data pada kolom ini tidak normal.

> - `CashbackAmount`
>   - Scaling : Menggunakan Min Max Scaler karena distribusi data tidak normal dan tidak terdapat extreme outliers.

```python
# Buat preprocessor menggunakan Pipeline
preprocessor = ColumnTransformer([
    ('Numerical 1', Pipeline([
        ('Imputer 1', SimpleImputer(strategy='median')),
        ('Scaler 1', RobustScaler())
    ]), ['Tenure', 'WarehouseToHome', 'NumberOfAddress', 'DaySinceLastOrder']),

    ('Numerical 2', Pipeline([
        ('Scaler 2', MinMaxScaler())
    ]), ['NumberOfDeviceRegistered', 'SatisfactionScore', 'CashbackAmount', 'Complain']),

    ('Categorical', Pipeline([
       ('Encoder', OneHotEncoder(handle_unknown='ignore'))
    ]), ['PreferedOrderCat', 'MaritalStatus'])
])

# Transformasi data train dan test
X_train = preprocessor.fit_transform(X_train_raw)
X_test = preprocessor.transform(X_test_raw)
```

```python
feature_names = preprocessor.get_feature_names_out()

# Mengubah X_train dan X_test kembali menjadi DataFrame (sudah diproses)
X_train = pd.DataFrame(X_train, columns=feature_names, index=y_train.index)
X_test = pd.DataFrame(X_test, columns=feature_names, index=y_test.index)
```

> Model Benchmark adalah model referensi yang digunakan sebagai tolak ukur untuk mengevaluasi kinerja model yang sedang dikembangkan.
> Dengan adanya benchmark, kita dapat menentukan apakah model baru memberikan hasil yang lebih baik dibandingkan dengan pendekatan sederhana atau model sebelumnya.

```python
# Definisikan model-model klasifikasi yang akan diuji
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(n_estimators=100, random_state=42)
}

results = []

# Loop untuk iterasi metode resampling
for resample_name, resample_method in resampling_methods.items():
    
    # Aplikasikan resampling
    if resample_method:
        X_resampled, y_resampled = resample_method.fit_resample(X_train, y_train)
    else:
        X_resampled, y_resampled = X_train, y_train
    
    # Loop untuk iterasi metode feature selection
    for fs_name, fs_method in feature_selectors.items():
        
        # Apply feature selection
        if fs_method:
            X_resampled_selected = fs_method.fit_transform(X_resampled, y_resampled)
            X_test_selected = fs_method.transform(X_test)
        else:
            X_resampled_selected = X_resampled
            X_test_selected = X_test

        # Loop  untuk iterasi model klasifikasi
        for model_name, model in models.items():

            # Train model
            model.fit(X_resampled_selected, y_resampled)
            y_pred = model.predict(X_test_selected)

            # Evaluasi dengan F2-Score
            f2 = fbeta_score(y_test, y_pred, beta=2)

            # Simpan hasil
            results.append({
                'Model': model_name,
                'Feature Selection': fs_name,
                'Resampling': resample_name,
                'F2 Score': f2
            })

results_df = pd.DataFrame(results).sort_values(by="F2 Score", ascending=False)
display(results_df)
```

> Berdasarkan hasil evaluasi terhadap berbagai model dengan metode resampling dan seleksi fitur yang berbeda, model XGBoost dengan SMOTE tanpa feature selection menunjukkan performa terbaik dengan nilai F2-score tertinggi.
>
> SMOTE (Synthetic Minority Over-sampling Technique) digunakan untuk meningkatkan jumlah sampel kelas minoritas, sehingga model dapat belajar dengan lebih baik tanpa bias terhadap kelas mayoritas.
>
> Ketika seleksi fitur diterapkan, performa model sedikit menurun, hal ini menunjukkan bahwa semua fitur yang tersedia memiliki kontribusi yang berarti, sehingga tidak perlu dilakukan penghapusan fitur.
>
> Dibandingkan dengan Random Forest dan Logistic Regression, XGBoost memberikan hasil yang lebih baik dalam generalisasi.
XGBoost memiliki banyak parameter yang dapat dituning, seperti learning rate, max depth, regularisasi, dan sampling, yang memungkinkan peningkatan performa lebih lanjut melalui optimasi hyperparameter.

```python
# Terapkan SMOTE untuk menangani class imbalance
smote = SMOTE(sampling_strategy=0.5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Inisialisasi model XGBoost SEBELUM tuning
xgb_before = XGBClassifier(n_estimators=100, random_state=42)
xgb_before.fit(X_resampled, y_resampled)

# Prediksi pada train dan test
y_train_pred_before = xgb_before.predict(X_train)
y_test_pred_before = xgb_before.predict(X_test)

# Hitung F2-score sebelum tuning
train_f2_before = fbeta_score(y_train, y_train_pred_before, beta=2)
test_f2_before = fbeta_score(y_test, y_test_pred_before, beta=2)

print("=== Performa Sebelum Tuning ===")
print(f"F2-Score (Train): {train_f2_before:.4f}")
print(f"F2-Score (Test): {test_f2_before:.4f}")

print("\nTrain Data Classification Report:")
print(classification_report(y_train, y_train_pred_before))
print("\nTest Data Classification Report:")
print(classification_report(y_test, y_test_pred_before))
```

## 4.2. Hyperparameter Tuning Menggunakan GridSearchCV

> Karena model ini telah menunjukkan performa terbaik, langkah selanjutnya Saya akan melakukan optimasi hyperparameter menggunakan GridSearchCV untuk mendapatkan konfigurasi parameter yang optimal.
>
> Hyperparameter tuning dilakukan untuk meningkatkan generalisasi model, menghindari overfitting, dan memastikan model bekerja optimal pada data ketidakseimbangan kelas. Dengan tuning, model XGBoost dapat lebih akurat dalam mendeteksi pelanggan churn tanpa kehilangan performa pada data baru.

```python
pipeline = ImbPipeline([
    ('resampling', SMOTE(sampling_strategy=0.5, random_state=42)),  # Best resampling method
    ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
])

# param_grid = {
#     'classifier__n_estimators': [280, 300],      # Small variation in trees  
#     'classifier__max_depth': [4, 5],             # Keeping depth low to avoid overfitting  
#     'classifier__learning_rate': [0.03, 0.035],  # Tiny tweak for stability  
#     'classifier__subsample': [0.88, 0.9],        # Minor adjustment for randomness  
#     'classifier__colsample_bytree': [0.4, 0.45], # Very small change for feature sampling  
#     'classifier__gamma': [0, 0.1],               # Keep regularization light  
#     'classifier__reg_alpha': [0.4, 0.5],         # Small change for L1 regularization  
#     'classifier__reg_lambda': [0.9, 0.95],       # Tiny adjustment for L2 regularization  
#     'classifier__scale_pos_weight': [3]          # Keeping it fixed to maintain class balance  
# }  

param_grid = {
    'classifier__n_estimators': [280, 300],      # Small variation in trees  
    'classifier__max_depth': [4, 5],             # Keeping depth low to avoid overfitting  
    'classifier__learning_rate': [0.03, 0.035],  # Tiny tweak for stability  
    'classifier__subsample': [0.88, 0.9],        # Minor adjustment for randomness  
    'classifier__colsample_bytree': [0.4, 0.45], # Very small change for feature sampling  
    'classifier__gamma': [0, 0.1],               # Keep regularization light  
    'classifier__reg_alpha': [0.5, 0.6],
    'classifier__reg_lambda': [0.9, 1.2],
    'classifier__scale_pos_weight': [3]          # Keeping it fixed to maintain class balance  
} 


# Define scoring metric for F2-score
f2_scorer = make_scorer(fbeta_score, beta=2)

# Perform Grid Search with execution time tracking
start_time = time.time()
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring=f2_scorer,
    cv=5,
    verbose=2,
    n_jobs=-1,
    return_train_score=True
)
grid_search.fit(X_train, y_train)

y_train_pred = grid_search.predict(X_train)
y_test_pred = grid_search.predict(X_test)

train_f2_score = fbeta_score(y_train, y_train_pred, beta=2)
test_f2_score = fbeta_score(y_test, y_test_pred, beta=2)

print(f"\nTrain Data F2-Score: {train_f2_score:.6f}")
print(f"Test Data F2-Score: {test_f2_score:.6f}")

end_time = time.time()
total_time = end_time - start_time
print(f'Total waktu fit: {total_time}')
```

      - Buat program loyalitas dengan reward berbasis waktu. Misalnya, pelanggan yang sudah berlangganan lebih dari 6 bulan mendapatkan diskon eksklusif.
      - Berikan insentif untuk perpanjangan keanggotaan seperti free shipping, diskon progresif, atau akses early bird ke produk baru.
  - `Complain` (Mengatasi Keluhan Pelanggan (Complain)) :

    **Sistem Kompensasi yang Adil**
      - Jika pelanggan mengalami masalah (contoh: keterlambatan pengiriman, produk cacat), tawarkan diskon atau penggantian produk tanpa biaya tambahan.
      - Tawarkan garansi kepuasan pelanggan, misalnya “Uang Kembali dalam 7 Hari” untuk meningkatkan kepercayaan pelanggan.

### 5.3.1. Model

Meskipun model telah baik dalam melakukan prediksi, ada beberapa poin yang masih bisa ditingkatkan untuk hasil yang lebih optimal.
- **Nilai precision masih rendah (66%)** → Model masih sering salah memprediksi pelanggan sebagai churn padahal tidak. Untuk mengembangkan model yang lebih baik, dapat fokus menemukan keseimbangan antara menangkap True Positive dan meminimalisir False Negative dan False Positive.
- **Menaikkan Threshold Tuning untuk menangani imbalance pada label** → Berdasarkan kurva AUC, model masih dapat diperbaiki dengan menyesuaikan threshold prediksi agar mendapatkan keseimbangan precision dan recall yang optimal. Coba naikkan threshold prediksi (misalnya dari 0.5 ke 0.6 atau 0.7) agar hanya pelanggan dengan probabilitas churn yang lebih tinggi yang diprediksi sebagai churn.
- **Model tidak bisa dibiarkan statis.** → Model constraints mengacu pada keterbatasan model dalam menyesuaikan diri dengan perubahan data dan pola pelanggan. Untuk mengatasi hal ini, model harus selalu diperbarui dan disesuaikan dengan data terbaru.
